```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/Cardiovascular_Disease_Dataset.csv")
```

```python
data.head(10)
```

```python
df=data.drop(['patientid'],axis=1)
```

```python
df.head()
```

```python
x=df[['age','gender','chestpain','restingBP','serumcholestrol','fastingbloodsugar','restingrelectro','maxheartrate','exerciseangia','oldpeak','slope','noofmajorvessels']]#independent
y=df['target']#dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
Knn=KNeighborsClassifier(n_neighbors=10)
```

```python
Knn.fit(x_train,y_train)
```

```python
neighbors_predictions=Knn.predict(x_test)
```

prediction for new person-

```python
new_data = {'age':25, 'gender':1,'chestpain':1,'restingBP':110,'serumcholestrol':250,'fastingbloodsugar':85,'restingrelectro':0,'maxheartrate':120,'exerciseangia':1,'oldpeak':3,'slope':1,'noofmajorvessels':2}

index=[1]
```

```python
my_data=pd.DataFrame(new_data,index)
```

```python
my_data
```

```python
new_predictions=Knn.predict(my_data)
```

```python
print("the patient will have",new_predictions,'ÇSV')
```

```python
from sklearn.metrics import recall_score
```

```python
score=recall_score(neighbors_predictions,y_test)
```

```python
print(score)
```

```python
from sklearn.metrics import f1_score
Fscore=f1_score(neighbors_predictions,y_test)
```
